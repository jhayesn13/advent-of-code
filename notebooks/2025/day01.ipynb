```python
rotations = []
for rot in inp_test_2:
    rotation = rot[1:]
    if rot.startswith('L'):
        rot_with_direction = '-' + rotation
    else:
        rot_with_direction = '+' + rotation
    rotations.append(rot_with_direction)
```
